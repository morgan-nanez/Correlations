```python
def Mbh2sigmaMM13(sigma):
    #MM13
    first_scatter = (np.random.normal(8.32, 0.05))
    second_scatter = (np.random.normal(5.64, 0.32))
    ans = np.random.normal((first_scatter+(second_scatter*log10(sigma/200))), 0.38)
    return 10**(ans)
```

```python
def Mbh2sigmaKH13(sigma):
    #KH13
    first_scatter = np.random.normal(0.309, 0.037)
    second_scatter = np.random.normal(4.48, 0.29)
    ans = (first_scatter * pow((sigma / 200), second_scatter))
    logM = log10(ans) 
    scatterM = np.random.normal(logM, 0.29) + 9
    return 10**scatterM
```
