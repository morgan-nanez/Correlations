```python
def Mk2mStar(mag):
    """
    converting from k-band luminosity to M* using Eq.2 from Ma et al. (2014)
    valid for early-type galaxies
    """
    Mstar = 10.58-0.44*(mag + 23)
    return 10**(Mstar)
```

```python
#mass total

def getMtot(ans, q):
    return ans / ((q/(1+q)**2)**(3/5))
```
